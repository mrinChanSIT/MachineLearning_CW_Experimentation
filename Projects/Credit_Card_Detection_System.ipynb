```python
#splitting dataset into features and target
X = ccdata.drop(columns="Class", axis = 1)
```

```python
Y = ccdata['Class']
```

```python
print(X.head())
```

```python
#splitting data into training and test dataset
```

```python
#using train_test_split funtion from sklearn
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2, stratify = Y, random_state= 2 )
```

```python
print(X.shape, x_test.shape, x_train.shape)
```

```python
#creating an object for the model - Logistic regression

#read more about - standard scaler - scalar.fit_transform - transform
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

print(x_train_scaled)

model = LogisticRegression()
model.fit(x_train_scaled, y_train)
```

```python
#model evaluation
```

```python
x_train_prediction = model.predict(x_train_scaled)
x_train_acc = accuracy_score(x_train_prediction, y_train)
print('Accuracy score on training data',x_train_acc)
```

```python
x_test_prediction = model.predict(x_test_scaled)
print(len(x_test_prediction))
print(y_test.info())
```

```python
x_test_acc = accuracy_score(x_test_prediction,y_test)
print('Accuracy score on test data - scaled',x_test_acc)
```
